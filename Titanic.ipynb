```python


# List of columns to calculate value counts for
columns_to_count = ['Sex', 'Pclass']  # Add any other columns you want

# Use groupby and value_counts for each column
result = {}
for col in columns_to_count:
    result[col] = df.groupby('Survived')[col].value_counts().unstack(fill_value=0)

# Display results for all columns
for col, table in result.items():
    print(f"Value counts for '{col}' grouped by 'Survived':\n")
    print(table)
    print("\n")
```

```python


# List of columns for which to calculate averages
columns_to_average = ['Age', 'Fare' ,"Sex" ,"Pclass"]  # Add more columns if needed

# Calculate averages grouped by 'Survived'
averages = df.groupby('Survived')[columns_to_average].mean()

# Display the result
print("Average values grouped by 'Survived':\n")
print(averages)
```
